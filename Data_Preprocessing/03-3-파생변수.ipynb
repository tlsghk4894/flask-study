```python
from hossam import load_data
import numpy as np
```

```python
origin = load_data("grade")
origin
```

```python
df1 = origin.copy()
df1['직접합계'] = df1['국어'] + df1['영어'] + df1['수학'] + df1['과학']
df1
```

```python
df2 = origin.copy()
df2['합계'] = df2[df2.columns[2:6]].sum(axis=1)
df2
```

```python
df3 = origin.copy()
df3['합계']=df3.iloc[:, 2:].sum(axis=1)
df3
```

```python
df3['평균'] = df3.iloc[:, 2:6].mean(axis=1)
df3['최대값'] = df3.iloc[:,2:6].max(axis=1)
df3['최소값'] = df3.iloc[:,2:6].min(axis=1)
df3['표준편차'] = df3.iloc[:,2:6].std(axis=1)
df3['분산'] = df3.iloc[:,2:6].var(axis=1)
df3
```

```python
df4 = origin.copy()
df4['합계'] = df4.iloc[:, 2:6].sum(axis=1)
df4['평균'] = df4.iloc[:, 2:6].mean(axis=1)
df4
```

```python
df4['결과1'] = np.where(df4['평균']>= 80, '합격', '불합격')
df4
```

```python
df4['결과2'] = "불합격"
df4.loc[df4['평균']>=80, "결과2"] ='합격'
df4
```

```python
conditions = [ (df4['평균'] >=90),
                (df4['평균'] >=80),
                (df4['평균'] >=70)]

value = ['A', 'B', 'C']

df4['학점1'] = np.select(conditions, value, default = 'F')
df4
```

```python
def getScore(x):
    if x>=90:
        score = 'A'
    elif x>=80:
        score = 'B'
    elif x>=70:
        score = 'C'
    else:
        score = 'F'

    return score
```

```python
df4['학점2'] = df4['평균'].apply(getScore)
df4
```

```python
df4['학점3'] = df4['평균'].apply(
    lambda x: {10: "A", 9: "A", 8: "B", 7: "C"}.get(x//10, "F"))
df4
```

```python
df4['성별코드'] = df4['성별'].map({"남자": "M", "여자": "F"})
df4
```

```python
origin = load_data("online_shop_sales")
origin
```

```python
df1 = origin.copy()
#고객별 총 구매 금액
총구매금액 = df1['Price'] * df1['Quantity']
df1['총구매금액'] = 총구매금액
df1
```

```python
각고객별총구매금액 = df1.groupby('UserID')['총구매금액'].sum()
각고객별총구매금액
```

```python
#총 구매 금액이 가장 높은 상위 2명의 고객을 찾아 UserID와 총 구매 금액을 출력
top2 = 각고객별총구매금액.sort_values(ascending=False).head(2)
top2
```

```python
#최종 결과 = 총 구매 금액을 기준으로 내림차순 정렬로 전제 데이터 제시
df1['각고객별총구매금액'] = df1['UserID'].map(각고객별총구매금액)
df_final = df1.sort_values(by='각고객별총구매금액', ascending=False)
df_final
```
