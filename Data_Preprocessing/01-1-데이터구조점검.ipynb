```python
from hossam import load_data        #데이터 로딩 함수
from pandas import CategoricalDtype #명목형 타입 클래스

origin = load_data("grade")
origin
```

```python
rows, cols = origin.shape
print("%d행, %d열의 %d차원 데이터" % (rows,cols,origin.ndim))
```

```python
origin.info()
```

```python
컬럼타입변경df = origin.copy()
컬럼타입변경df.info()
```

```python
컬럼타입변경df['국어'] = 컬럼타입변경df['국어'].astype('float')
컬럼타입변경df.info()
```

```python
타입변환df = origin.astype({'국어':'float', '영어':'float',
                        '수학':'str','과학':'bool'})
타입변환df.info()
```

```python
origin.info()
```

```python
컬럼단위_명목df = origin.copy()
컬럼단위_명목df['학년'] = 컬럼단위_명목df['학년'].astype('category')
컬럼단위_명목df['성별'] = 컬럼단위_명목df['성별'].astype('category')
컬럼단위_명목df.info()
```

```python
범주형df = origin.astype({'학년': 'category', '성별':'category'})
범주형df.info()
```

```python
컬럼단위_서열척도df = origin.copy()
컬럼단위_서열척도df['학년'] = 컬럼단위_서열척도df['학년'].astype(
    CategoricalDtype(categories=[1,2,3,4], ordered=True))
컬럼단위_서열척도df.info()
```

```python
서열척도df = origin.astype({
    "학년": CategoricalDtype(categories=[1,2,3,4], ordered=True)})
서열척도df.info()
```

```python
이름변경df = origin.rename(
    columns = {'국어':'kor', '영어':'eng', '수학':'math', '과학':'sci'},
    index = {'석영': 'suk','호영':'ho','민철':'min','수현':'su'}
)
이름변경df
```

연습문제 2 데이터 구성하기


```python
# 네 종류의 식단 A1, A2, A3, A4에 따른 식이용법이 혈액응고시간에 영향을 주는지 알고자 실험한 자료가 다음과 같다
# 인자의 수준을 나타내는 변수를 class로, 실습값을 time으로 하여 데이터를 구성하라
# class 변수는 명목형 변수이다.
data = [
    {'class': 'A1', 'time': 62},{'class': 'A1', 'time': 60},{'class': 'A1', 'time': 63},{'class': 'A1', 'time': 59},{'class': 'A1', 'time': 61},
    {'class': 'A2', 'time': 62},{'class': 'A2', 'time': 67},{'class': 'A2', 'time': 71},{'class': 'A2', 'time': 64},{'class': 'A2', 'time': 65},{'class': 'A2', 'time': 66},
    {'class': 'A3', 'time': 68},{'class': 'A3', 'time': 66},{'class': 'A3', 'time': 71},{'class': 'A3', 'time': 67},{'class': 'A3', 'time': 68},{'class': 'A3', 'time': 69},{'class': 'A3', 'time': 70},
    {'class': 'A4', 'time': 58},{'class': 'A4', 'time': 62},{'class': 'A4', 'time': 60},{'class': 'A4', 'time': 59},{'class': 'A4', 'time': 61},{'class': 'A4', 'time': 63},{'class': 'A4', 'time': 64}
]
import pandas as pd
df = pd.DataFrame(data)
df['class'] = df['class'].astype('category')
df.info()
```
