```python
from hossam import load_data
```

```python
origin = load_data('grade')
origin
```

```python
단일정렬df = origin.sort_values('학년', ascending=False)
단일정렬df
```

```python
복수정렬df = origin.sort_values(['학년','국어'],
                            ascending=[False,False])
복수정렬df
```

```python
#인덱스로 정렬된거임 (현재 인덱스는 이름)
인덱스정렬df = origin.sort_index(ascending=False)
인덱스정렬df
```

```python
# r1 = origin[origin['국어'] > 80] 과 같은 표현
r1 = origin.query('국어 > 80')
r1
```

```python
#r2 = origin[origin["성별"]=='남자']
r2 = origin.query("성별 == '남자'")
r2
```

```python
#r3 = origin[(origin['국어'] >80) & (origin['수학'] >80)]
r3 = origin.query('국어 >80 and 수학 > 80')
r3
```

```python
#r4 = origin[(origin['국어']<=80) | (origin['수학'] <=80)]
r4 = origin.query('국어 <= 80 or 수학<=80')
r4
```

```python
point = int(input('점수를 입력하세요:')) #82 입력
print("국어 점수가 %d점 보다 높은 학생" %point)

r5 = origin.query('국어 > @point')
r5
```

VIP 고객을 찾아라! 



```python
origin = load_data('vip_customer')
origin.head()
```

```python
우수서울고객 = origin.query('TotalSpending > 100000 and City == "서울"')
우수서울고객
```

```python
최종vip5명 = 우수서울고객.sort_values('TotalSpending', ascending=False).head(5)
최종vip5명
```

```python
#원하는 열만 추출
보고할내용 = 최종vip5명.filter(['Name', 'City', 'TotalSpending'])
보고할내용
```
