```sql
create table subway_test(
    id int not null auto_increment comment '일련번호',
    type_of_track varchar(50) not null comment '선별',
    peak int not null default 0 comment '첨두(피크시간 사람 수)',
    best_peak int not null default 0 comment '최고 사람 수',
    most_crowded varchar(50) not null default 0 comment '최대 혼잡구간',
    primary key(id)
) default charset utf8 collate utf8_bin comment '수도권 전철 혼잡도'
```
