# [LAB-02] Pandas - 5. 데이터 내보내기

## 라이브러리 참조

```python
from hossam import load_data
```

```python
origin = load_data('grade')
origin
```

```python
origin.to_excel("1.xlsx")
```

```python
origin.to_excel("2.xlsx",
                index=True, # 행 인덱스 포함 여부
                index_label="Name" # 행 인덱스 열 이름 지정
                )
```

```python
origin.to_excel("3.xlsx",
                index=False, # 행 인덱스 미포함
                columns=['학년', '성별','국어'], # 내보낼 열 지정
                header=['level','sex','kor'] # 열 이름 지정
                )
```

```python
origin.to_csv("1.csv", encoding='utf-8')
```

```python
origin.to_csv("2.csv",
              encoding="utf-8", # 인코딩 지정
              index=True, # 행 인덱스 포함 여부
              index_label="Name" # 행 인덱스 열 이름 지정
              )
```

```python
origin.to_csv("3.csv",
              encoding="utf-8",
              index=False, # 행 인덱스 미포함
              columns=['학년','성별','국어'], # 내보낼 열 지정
              header=['level','sex','kor'] # 열 이름 지정
              )
```
