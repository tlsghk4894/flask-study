```python
exam = np.array([60,83,75,92,86])

#첫 번째 파라미터로 주어진 조건이 참일 경우 '합격'
#거짓일 경우 '불합격' 문자열을 반환
result = np.where(exam>=80, '합격','불합격')
result
```

```python
exam = np.array([60,85,51,79,94,88])

conditions = [(exam >=90),
              (exam >=80),
              (exam>=70)]

value = ['A','B','C']

result = np.select(conditions, value, default='F')
result
```

```python
score = [72.5, 88.0, 91.5, 67.0, 80.0, 59.5]

ndarray_score = np.array(score)
ndarray_score
print(ndarray_score[ndarray_score>=80.0])
```

```python
conditions = [(ndarray_score >=90),
              (ndarray_score >=80),
              (ndarray_score>=70)]
result = np.select(conditions, value, default='F')
print("원본 점수 :", ndarray_score)
print("80점 이상:", ndarray_score[ndarray_score>=80.0] )
result
```

2차원 배열 ndarray로 변환

```python
import pandas as pd
import numpy as np

myarray = np.array([
    [98,72,80,64],
    [88,90,80,72],
    [92,88,82,76]
])

print(myarray)

print("차원크기:", myarray.ndim)
print("차원별 원소수:", myarray.shape)
print("원소의 데이터 타입:", myarray.dtype)
```

```python
bool_array = np.array([
    [True, False, True, False],
    [True, True, True, False],
    [True,True,True,False]
])

# 조건에 맞는 항목만 1차 배열로 추출
# myarray는 이전 실습에서 생성한 2차원 배열
result1 = myarray[bool_array]
result1
```
