### 배열의 크기

#### `len()` 함수를 사용하여 배열의 크기를 확인할 수 있다.

```python
arr = np.array([10,30,50,70,90])
size = len(arr)
print("배열의 원소는 %d개 입니다." % size)
```

```python
mydata = np.array([5,10,15,20,25])

print(mydata[0])
print(mydata[1])
print(mydata[3])

print(mydata[1:4])
print(mydata[-4:-1])

```

```python
mydata = np.array([100,110,120,130,140])

for i,v in enumerate(mydata):
    print("%d번째 원소 => %d" % (i,v))
    
```
