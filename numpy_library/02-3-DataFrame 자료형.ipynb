```python
from pandas import DataFrame, read_csv, read_excel
```

2차원 리스트를 DataFrame으로 변환

```python
grade_data = [
    [1,"남자",98,77,88,64],
    [2,"여자",88,90,62,72],
    [1,"남자",92,70,83,79],
    [3,"여자",63,60,31,70],
    [4,"남자", 75,50,90,88]
]

df = DataFrame(grade_data)
```

```python
df
```

인덱스와 컬럼 이름 지정

```python
# 인덱스 이름
i_names = ['철수','영희','민철','수현','호영']

#컬럼 이름 
c_names = ['학년','성별','국어','영어','수학','과학']

df = DataFrame(grade_data, index=i_names, columns = c_names)
df
```

```python
grade_dic = {
    '학년':[1,2,1,3,4],
    '성별':['남자','여자','남자','여자','남자'],
    '국어':[98,88,92,63,75],
    '영어':[77,90,70,60,50],
    '수학':[88,62,83,31,90],
    '과학':[64,72,79,70,88]
}

df = DataFrame(grade_dic)
df
```

인덱스와 인덱스 이름을 지정한 DataFrame

```python
df = DataFrame(grade_dic, index=i_names)

df.index.name = '이름'
df
```

딕셔너리를 원소로 갖는 리스트를 활용한 DataFrame 생성

```python
grade_list =[
    {"이름":"철수","학년":1,"성별":"남자","국어":98,"영어":77,"수학":88,"과학":64},
    {"이름":"영희","학년":2,"성별":"여자","국어":88,"영어":90,"수학":62,"과학":72},
    {"이름":"민철","학년":1,"성별":"남자","국어":92,"영어":70,"수학":83,"과학":79},
    {"이름":"수현","학년":3,"성별":"여자","국어":63,"영어":60,"수학":31,"과학":70},
    {"이름":"호영","학년":4,"성별":"남자","국어":75,"영어":50,"수학":90,"과학":88}
]

df = DataFrame(grade_list)
df
```

이미 생성된 DataFrame의 특정 컬럼을 index로 지정하기

```python
df2 = df.set_index('이름')
df2
```

복사본을 생성하지 않고 원본에 즉시 반영 -> 비추천

```python
df.set_index('이름', inplace=True)
df
```

인덱스 해제하기 (일반 컬럼으로 되돌린 복사본 리턴)

```python
df3 = df.reset_index()
df3
```

한번 더 사용시 일련번호 인덱스를 컬럼으로 돌려버림

```python
df3.reset_index()
```

인덱스 해제하면서 기존 인덱스 제거

```python
df3.reset_index(drop=True)
```
