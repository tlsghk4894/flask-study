```python
from hossam import load_data,load_info
```

```python
origin = load_data('GRADE')
origin
```

```python
#데이터 차원과 크기
print('차원:', origin.ndim)
print('크기:', origin.shape)
```

```python
#각 필드별 데이터 타입
origin.dtypes
```

```python
# 각 필드별 정보 확인 *필수
origin.info()
```

```python
#데이터 프레임은 열 다음 행임
origin['국어']
```

```python
type(origin['국어'])
```

```python
#대괄호를 이중으로 사용하여 원하는 열만 필터링 하여 데이터 프레임 생성
df2 = origin[['국어','영어','수학']]
df2
```

```python
#특정 행 데이터 추출
print(origin.loc['철수'])
print(type(origin.loc['철수']))
```

```python
#특정 위치의 데이터 추출, 반복문 쓸 때 편함 데이터 프레임은 반복문 별로 안씀
origin.iloc[0]  
```

```python
origin.iloc[1:4]  #1~3번째 행까지 추출
```

```python
#단일 값 확인하기 열->행 순서로 접근
origin['국어']['철수']
```

```python
# 행-> 열 순서로 접근
origin.loc['철수','국어']
```

```python
#인덱스 번호 기반 데이터 접근 행->열
origin.iloc[1,2]
```

```python
origin.iloc[:,3] #모든 행의 3번째 열 데이터
```

```python
# 값의 변경은 행->열 순으로 접근하는 경우만 가능함
origin.loc['철수','국어'] = 100
origin
```

```python
#상위 n건 확인
origin.head(2)
```

```python
#하위 n건 확인
origin.tail(2)
```
