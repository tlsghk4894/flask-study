## #01. Series 자료형
 
### 라이브러리 참조

```python
import numpy as np
from pandas import Series
```

```python
my_list = [10,30]
s1 = Series(my_list)
s1
```

```python
my_array = np.array([100,200])
s2 = Series(my_array)
s2
```

```python
my_dict = {"hello": 123, "world": 456, "nice": 789}
s3 = Series(my_dict)
s3
```

```python
my_series=Series({"hello": 123, "world":456,"nice":789})
my_series.dtype
```

```python
my_series.info()
```

```python
my_series = Series([10,20,30,40,50])
print("== 인덱싱 ==")
print(my_series[0])
print("== 슬라이싱 ==")
print(my_series[1:3])
```
