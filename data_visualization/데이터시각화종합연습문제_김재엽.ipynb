```python
# 자동차의 평균 연비 = (고속도로 연비(hwy) + 도심 연비(cty))) / 2
# 그걸 확인하기 위해 전체 자동차에 대한 평균 연비 먼저 구하기 => 모든 자동차의 연비에 대한 평균
# 그 값보다 높은 연비를 갖는 자동차는 Pass 아니면 Fail로 구분

# 1. 일단 전체 자동차에 대한 평균 연비 구하기
df9 = origin.copy()
전체_평균_연비 = ((df9['hwy'] + df9['cty'])/2).mean()
전체_평균_연비
```

```python
# 2. 각 자동차별 평균 연비를 구하기
df9['평균 연비'] = (df9['hwy'] + df9['cty']) / 2
df9
```

```python
# 3. 전체 평균 연비와 비교하여 Pass/Fail 부여
df9['연비 등급'] = np.where(df9['평균 연비'] >= 전체_평균_연비, 'Pass', 'Fail')
df9
```

```python
# 문제 1. Pass 등급과 Fail 등급의 비율을 시각화하라.
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 1280             #그래프 가로 크기
height_px = 720             #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 비율 시각화
sb.countplot(data=df9, x='연비 등급')

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid()

# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```

```python
# 문제 3. Pass 등급을 받은 자동차들에 대해 연비가 높은 순으로 각각의 연비를 시각화하라(전처리 과정에서 발생하는 결측치는 모두 삭제하고 같은 모델에 대해서는 평균값을 사용한다.)
# 결측치가 존재하나요? 
# df11 = origin.copy()
# print(df11.isnull().sum())
# -> ㄴㄴ 없음
# 그럼 같은 모델에 대해서는 평균값 사용하기만 진행, 일단 Pass 등급을 받은 자동차들만 필터링하기, df9는 위에서 연비 등급을 부여한 프레임 사용
df11 = df9[df9['연비 등급']=='Pass']
# df11
# 같은 모델에 대해 평균 연비의 평균 값 사용
df12 = df11.groupby('model')['평균 연비'].mean().reset_index()
df12
```

```python
# 연비가 높은 순으로 시각화 해야하니까 정렬 먼저 진행
df12=df12.sort_values('평균 연비', ascending=False)
df12
```

```python
# 시각화 진행
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 2400             #그래프 가로 크기
height_px = 1200             #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) barplot 그리기
sb.barplot(
    data=df12,
    x="model",
    y="평균 연비",
)

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid()

# 4) 출력
plt.tight_layout()                      #여백 제거
plt.xticks(rotation=45)
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```
