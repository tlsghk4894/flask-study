전국 실업률 분포 변화
- unemployment_age 데이터는 2000년부터 2022년까지 행정구역(시도)/연령별 실업률을 담고 있다.
- 이 데이터를 토대로 지역별 실업률이 어떻게 변화하고 있는지 확인해보자

```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
from pandas import melt, pivot_table
import seaborn as sb
import numpy as np
```

```python
my_dpi = 200
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
my_font_name = "Noto Sans KR"

plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 4
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data('unemployment_age')
origin.head()
```

ffill() 함수를 통해 그 전에 있던 값으로 채워넣기

```python
df = origin.copy()
df['시도별'] = df['시도별'].ffill()
df.head(10)
```

년도에 따른 전국 평균 실업률 변화
- 데이터 전처리 - 데이터 재구조화

```python
df2 = melt(df, id_vars=['시도별', '연령계층별'], var_name ='년도', value_name='실업률')
df2.head()
```

시/도에 따른 년도별 평균 실업률

```python
tdf1 = df2[['시도별','년도','실업률']].groupby(['시도별','년도'], as_index=False).mean()
tdf1
```

전국에 대한 년도별 평균 실업률

```python
tdf2 = tdf1[['년도','실업률']].groupby('년도').mean()
tdf2
```

데이터 시각화

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 2000            #그래프 가로 크기
height_px = 800            #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 -> seaborn 사용
# sb.boxplot(data=df1, orient='h')
sb.lineplot(data=tdf2, x=tdf2.index, y="실업률")

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid(True)
ax.set_title("년도에 따른 전국 평균 실업률 변화(전국평균)", fontsize=12, pad=8)


# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```

- 2000년 이후 증감을 반복하는 모습을 보이지만 전체적으로 증가하는 추세
- 2020년 이후 급격히 감소함

년도에 따른 연령대별 전국 평균 실업률 변화
- 앞에서 만든 데이터 재구조화 결과에 이어서 진행

```python
gdf = df2.filter(['년도', '연령계층별','실업률']).groupby(['년도','연령계층별'], as_index=False).mean()
gdf
```

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 2000            #그래프 가로 크기
height_px = 800            #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 -> seaborn 사용
# sb.boxplot(data=df1, orient='h')
sb.lineplot(data=gdf, x='년도', y="실업률", hue='연령계층별')

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid(True)
ax.set_title("년도에 따른 전국 평균 실업률 변화(연령대별)", fontsize=12, pad=8)


# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 3600            #그래프 가로 크기
height_px = 1200            #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 2                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, [ax1,ax2] = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 -> seaborn 사용
# sb.boxplot(data=df1, orient='h')
sb.lineplot(data=tdf2, x=tdf2.index, y="실업률", ax=ax1)
sb.lineplot(data=gdf, x='년도', y="실업률", hue='연령계층별', ax=ax2)

fig.suptitle("년도에 따른 전국 평균 실업률 변화", fontsize=14, color='#006600')

fig.subplots_adjust(wspace=0.2, hspace=0.2)

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax1.title.set_text("전국평균")
ax1.title.set_fontsize(12)
ax1.grid()

ax2.title.set_text("연령대별 평균")
ax2.title.set_fontsize(12)
ax2.grid()

# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```
