```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
from pandas import pivot_table
import seaborn as sb
```

```python
my_dpi = 200
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
my_font_name = "Noto Sans KR"
my_theme = "dark"

sb.set_theme(style=my_theme)
plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 6
plt.rcParams['axes.unicode_minus'] = False
```

```python
origin = load_data("flights")
origin.head()
```

```python
df = origin.copy()
df['month'] = df['month'].map({
    "January":1, "February":2, "March":3, "April":4,
    "May":5, "June":6, "July":7, "August":8,
    "September":9, "October":10, "November":11, "December":12
})

df.head()
```

```python
df2 = pivot_table(df, index="year", columns="month", values="passengers")
df2
```

Heatmap 시각화
- 데이터의 패턴, 특히 밀도와 분포를 빠르게 파악할 수 있게 해주는 시각적 도구
- 행과 열을 가진 행렬 형태의 데이터를 색상으로 나타내어 각 셀의 값에 따라 색상이 변한다.

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 1000             #그래프 가로 크기
height_px = 1000             #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) heatmap 그리기
#-> annot = True : 수치값을 함께 표시
#-> fmt="0.1f" : 소수점 첫째자리까지 표시 , annot이 True일 때만 적용
# -> linewidths=0.5 : 셀 간격 선 두께
# -> cmap="PuOr" : 색상표 지정(색상표는 seaborn에서 제공하는 것 중 선택 가능)
# cmap 종류 : Greys, Purples, Blues, Greens, Oranges, Reds,
#            BuGn, BuPu, GnBu, OrRd, PuBu, YlGn, YlOrBr, YlOrRd,
#            RdPu, PuRd, OrPu, RdYlBu, RdYlGn, Spectral, coolwarm, bwr, seismic, vanimo, PiYG, PRGn, BrBG, PuOr
sb.heatmap(data=df2, annot = True, fmt="0.1f", linewidths=0.5, cmap="PuOr")

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.set_xlabel("")
ax.set_ylabel("")
ax.xaxis.tick_top()                    #x축의 변수 이름을 상단으로 이동

# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```

- 여름(6-8)월에 진한 색-> 승객수가 가장 많음
- 년도가 증가할수록 전체 색 농도가 짙어짐 -> 항공 운송 수요 증가 추세
- 연도별 패턴은 거의 동일하지만 크기만 증가 -> 계절성 + 추세 구조가 혼합된 전형적 시계열 패턴
