```python
from hossam import load_data
from matplotlib import pyplot as plt
from matplotlib import font_manager as fm
import seaborn as sb
```

데이터 불러오기

```python
origin = load_data("traffic_acc")
origin.head()
```

```python
df = origin.drop('월', axis=1).groupby('년도').sum()
df
```

그래프 초기화

```python
my_dpi = 200
font_path = "./NotoSansKR-Regular.ttf"
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
font_name = font_prop.get_name()
my_font_name = "Noto Sans KR"
my_theme = "dark"

sb.set_theme(style=my_theme)
plt.rcParams['font.family'] = my_font_name
plt.rcParams['font.size'] = 10
plt.rcParams['axes.unicode_minus'] = False
```

시각화 기본 코드

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 1280             #그래프 가로 크기
height_px = 720             #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 -> seaborn 사용
# sb.boxplot(data=df1, orient='h')
sb.lineplot(data=df, x=df.index, y='발생건수', 
            color="#0066ff", linewidth=0.7, linestyle=':', marker="o", markersize=2,
            markerfacecolor="#00ff00", markeredgecolor="#ff0000", markeredgewidth=1)

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid(True)
# ax.set_title("😊", color="😊",fontsize=12, pad=8)
# ax.set_xlabel("😊",color="😊" ,fontsize=8, labelpad=5)
# ax.set_ylabel("😊", color="😊",fontsize=8, labelpad=5)
# ax.set_xlim([😊, 😊])
# ax.set_ylim([😊, 😊])

# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```

hue 파라미터 이해하기
- melt를 적용할 경우 index 해제

```python
df1 = df.reset_index()
df1
```

```python
#melt 함수란?
# 특정 열을 기준으로 나머지 열들을 행으로 녹이는(길게 만드는) 함수
# 형식: pd.melt(데이터프레임, id_vars=기준열, value_vars=녹일열들, var_name=녹인열이름)
df2 = df1.melt(id_vars='년도', value_vars=['발생건수', '사망자수','부상자수'],var_name='구분')
df2
```

hue 파라미터를 사용한 시각화

```python
# 1) 그래프 초기화 (캔버스(fig)와 도화지(ax) 준비하기)
width_px = 1280             #그래프 가로 크기
height_px = 720             #그래프 세로 크기
rows = 1                    #도화지의 행 수
cols = 1                    #도화지의 열 수
figsize = (width_px / my_dpi, height_px / my_dpi)
fig, ax = plt.subplots(rows, cols, figsize = figsize, dpi = my_dpi)

# 2) 그래프 그리기 -> seaborn 사용
# sb.boxplot(data=df1, orient='h')
sb.lineplot(data=df2, x="년도", y='value', hue='구분')

# 3) 그래프 꾸미기 -> 도화지(ax)에 직접 적용
ax.grid(True)
# ax.set_title("😊", color="😊",fontsize=12, pad=8)
# ax.set_xlabel("😊",color="😊" ,fontsize=8, labelpad=5)
# ax.set_ylabel("😊", color="😊",fontsize=8, labelpad=5)
# ax.set_xlim([😊, 😊])
# ax.set_ylim([😊, 😊])

# 4) 출력
plt.tight_layout()                      #여백 제거
# plt.savefig("myplot.png", dpi=my_dpi)   #생략 가능
plt.show()                              #그래프 화면 출력
plt.close()                             #그래프 작업 종료
```
