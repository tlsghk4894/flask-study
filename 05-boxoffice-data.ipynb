반복문으로 데이터 요청 처리하기 (전체 결과 가져오기)

```python
import requests
from pandas import DataFrame

query = "파이썬"
page=1
size=50

# 반복문 돌린 결과를 저장하기 위한 리스트 
resultset= []

with requests.Session() as session:
    session.headers.update({ "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36","Authorization":"KakaoAK %s" % API_KEY})

    for i in range(1,51):
        # 바이너리(이진형식) 데이터 가져옴
        r = session.get(url, params={"query" : query, "page":i,"size":size})

        if r.status_code !=200:
            msg = "[%d Error] %s 에러가 발생함" %(r.status_code, r.reason)
            # raise Exception(msg) 이건 에러 발생하면 중간에 중단됨
            print(msg)              # 우리는 에러 발생해도 계속 수집해야하므로 에러 출력만 하고 continue
            continue

        print("%d page 데이터 수신 완료 >> %s" %(i,r))
                
        r.encoding ="utf-8"
        mydict = r.json()
        resultset.extend(mydict['documents'])

print("총 %d건의 데이터가 수집됨" % len(resultset))
```

```python
df = DataFrame(resultset)
df.to_excel("카카오 책 검색(%s)-전체 데이터.xlsx" % query)
```

파일 다운로드 함수 정의

```python
def download(url, target):
    session.headers.update({ "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"})
    
    try:
        r = session.get(url, stream=True)
        r.encoding = "utf-8"

        with open(target,"wb") as f:
            f.write(r.raw.read())
            print(target, "가 저장되었습니다.")
    except Exception as e:
        print(target, "저장 실패", e)
```

반복문을 통해 수집된 결과 내의 표지 이미지 다운로드 받기

```python
import os
import datetime as dt

dirname = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
os.mkdir(dirname)

for i, v in enumerate(resultset):
    file_path = os.path.join(dirname, "%d.jpg" %i)

    download(v['thumbnail'], file_path)
```

```python
# 비동기 처리 기능을 제공하는 모듈
from concurrent import futures
# 다운로드 결과가 저장될 폴더 생성하기
dirname = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
os.mkdir(dirname)

# 비동기 작업 내에서 이미지 다운로드 받기
with futures.ThreadPoolExecutor() as executor:
    #수집 데이터 내에서 반복문 수행
    for i, v in enumerate(resultset):
        #저장될 파일 경로 문자열 생성
        file_path = os.path.join(dirname, "%d.jpg" % i)
        #비동기 다운로드 요청
        executor.submit(download, v['thumbnail'], file_path)
```

```python
import requests
import json
import datetime as dt
from pandas import DataFrame

#날짜 변수 생성 (오늘 날짜로부터 1년전 날짜)
now = dt.datetime.now()
today = now.strftime("%Y-%m-%d")
startDateDelta = now - dt.timedelta(days=365)
startDate = startDateDelta.strftime("%Y-%m-%d")

data = {
    "startDate" : startDate,
    "endDate" : today,
    "timeUnit": "date",
    "keywordGroups": [
        {
            "groupName" : "Benz", #검색 주제
            #주제와 관련된 검색어
            "keywords": ["벤츠","e클래스","s클래스","c클래스","벤츠 cls"]
        },
        {
            "groupName" : "BMW", #검색 주제
            #주제와 관련된 검색어
            "keywords": ["BMW","5시리즈","x5","x7","x3"]
        },
        {
            "groupName" : "AUDI", #검색 주제
            #주제와 관련된 검색어
            "keywords": ["아우디","a6","e트론","a7","q5"]
        }
    ]
}
```
