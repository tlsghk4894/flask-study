```python
import pandas as pd

# 데이터 로드 
file_name = "../data/서울특별시_전월세가_2024.csv"

try:
    df = pd.read_csv(file_name, encoding='euc-kr')
except UnicodeDecodeError:
    df = pd.read_csv(file_name, encoding='utf-8')

print("데이터 로드 완료. 총 계약 건수:", len(df))
```

```python
# 데이터 필터링 

# 전월세구분이 월세인 거래만 필터링
df_wolse = df[df['전월세구분'] == '월세'].copy()
print(f"월세 거래 건수: {len(df_wolse)}건")

# 계약일을 datetime 형식으로 변환 YYYYMMDD 형식
df_wolse['계약일'] = pd.to_datetime(df_wolse['계약일'], format='%Y%m%d')
```
