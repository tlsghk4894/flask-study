서울 지하철 주소 구하기

파싱이 이상하게 돼서 찾아낸 판다스로 값 가져오기..... 판다스는 신이야
- 도로명주소만 톡 떼서 가져오기

```python
import pandas as pd
data=pd.read_csv('./서울교통공사_역주소 및 전화번호_20250318.csv', encoding='euc-kr')
# data.info()
# data.value_counts()
doro_addr = data['도로명주소'].tolist()
print(doro_addr)
```

위경도 변환

```python
import requests	

lo_api_key = "A0586DE0-1EC6-3305-86E6-39FD5161FDE5"
apiurl = "https://api.vworld.kr/req/address?"
x=[]
y=[]
try:
    for i in doro_addr:
        params = {
            "service": "address",
            "request": "getcoord",
            "crs": "epsg:4326",
            "address": i,
            "format": "json",
            "type": "road",
            "key": lo_api_key
        }
        response = requests.get(apiurl, params=params)
        x.append(response.json().get('response', {}).get('result', {}).get('point', {}).get('x'))
        y.append(response.json().get('response', {}).get('result', {}).get('point', {}).get('y'))
        # print(x,y)
        if response.status_code == 200:
            print(response.json())
        else:
            print("Error:", response.status_code)
except Exception as e:
    print("Exception occurred:", str(e))
```

csv파일에 x,y 열 추가 전에 데이터 확인

```python
for i in range(len(x)):
    if x[i] is None or y[i] is None:
        print(f"x좌표: {i,x[i]}, y좌표: {i,y[i]}")

# None 값 인덱스 저장
non = []
for i in range(len(x)):
    if x[i] is None or y[i] is None:
        non.append(i)

if len(x) == len(doro_addr) and len(y) == len(doro_addr):
    print("일치")
```

csv파일로 저장
- pandas에서 지원하는 함수 loc[행 인덱싱 값, 열 인덱싱 값] 사용

- 원본 데이터와 길이가 동일한 것을 확인한 뒤 진행

```python
path = './서울교통공사_역주소 및 전화번호_20250318.csv'
data = pd.read_csv(path, encoding='euc-kr')
for i in range(len(x)):
    data.loc[i, 'latitude'] = x[i]
    data.loc[i, 'longitude'] = y[i]

y_0 = 126.972554606456      #인터넷을 통해 위경도 값 하드코딩 (None 값 처리)
x_0 = 37.5571594835988
y_258 = 127.147008
x_258 = 37.550442

data.loc[0,'latitude'] = y_0
data.loc[0,'longitude'] = x_0
data.loc[258,'latitude'] = y_258
data.loc[258,'longitude'] = x_258
data.to_csv('./subway_stations_with_coords.csv', index=False, encoding='euc-kr')
data
```
